# Analysis and Visualizations

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=False)
cancelled_perc
```

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
cancelled_perc
```

```python
## 37% of reservations are cancelled
```

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize=(5, 4))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled', 'Canceled'],df['is_canceled'].value_counts())
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,4))
A1 = sns.countplot(x='hotel',hue='is_canceled',data=df)
plt.title('Reservation Status in different hotel')
plt.xlabel('hotel')
plt.ylabel('reservation number')
```

```python
RH=df[df['hotel']=='Resort Hotel']
RH['is_canceled'].value_counts(normalize = True)
```

```python
CH=df[df['hotel']=='City Hotel']
CH['is_canceled'].value_counts(normalize = True)
```

```python
Resort_hotel=RH.groupby('reservation_status_date')[['adr']].mean()
City_hotel=CH.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (20,8))
plt.title('Avg. Daily Rate in Hotel')
plt.plot(Resort_hotel.index, Resort_hotel['adr'],label='Resort Hotel')
plt.plot(City_hotel.index, City_hotel['adr'],label='City Hotel')
plt.legend(fontsize=20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month

plt.figure(figsize=(16, 8)) 
axis = sns.countplot(x='month', hue='is_canceled', data=df)
plt.title('Reservation Status per Month')
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.show()
```

```python
plt.figure(figsize=(16, 8)) 
sns.barplot(x='month',y='adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.title('ADR per month')

plt.show()
```

```python
##higher price lead to high percentage of cancelation
```

```python
canceled_data = df[df['is_canceled']==1]
top_5_country = canceled_data['country'].value_counts()[:5]
plt.figure(figsize= (8,8))
plt.title('Top 5 countries with reservation cancelation')
plt.pie(top_5_country , autopct = '%.2f',labels= top_5_country.index)
plt.show()
```
